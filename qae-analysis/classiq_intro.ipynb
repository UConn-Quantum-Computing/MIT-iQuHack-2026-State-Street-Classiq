```python
import classiq
classiq.authenticate()
```

```python
from classiq import *
```

```python
@qfunc
def main(x: Output[QNum], y: Output[QNum]) -> None:
    allocate(3, x)
    hadamard_transform(x)
    y |= x**2 + 1
```

```python
qmod = create_model(main)
```

```python
qprog = synthesize(qmod)
```

```python
show(qprog)
```

```python
job = execute(qprog)
```

```python
pc = job.get_sample_result().parsed_counts
print(pc)
```
